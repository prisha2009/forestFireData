```python
import numpy
import matplotlib.pyplot as mt
import pandas
```

```python
data = pandas.read_csv("amazon.csv")
data.shape
```

```python
# data.head(10)
data = data.replace(0,numpy.nan)
filterData = data.dropna(subset=["number"])
filterData
```

```python
unMonths = list(filterData.month.unique())
ffire = filterData.groupby("month")["number"].sum()
ffire = ffire.reindex(unMonths,axis=0)
ffire = ffire.to_frame()
ffire.reset_index(level=0, inplace=True)
```

```python
mt.figure(figsize=(20,25))

mt.bar(
    ffire["month"],
    ffire["number"],
    color = [0.5,0.1,0.5,0.6]
)
mt.show()
```
